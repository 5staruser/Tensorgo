```python
from rouge_score import rouge_scorer

def calculate_rouge(reference, summary):
    # Initialize the ROUGE scorer
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, summary)
    return scores

# Example usage
reference_text = text
summarized_text = ext_summary

rouge_scores = calculate_rouge(reference_text, summarized_text)
print("ROUGE-1:", rouge_scores['rouge1'])
print("ROUGE-2:", rouge_scores['rouge2'])
print("ROUGE-L:", rouge_scores['rougeL'])
```

```python
import whisper
import moviepy.editor as mp
import os
import gradio as gr
from summarizer import Summarizer
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast

# Load the Whisper model
whisper_model = whisper.load_model("base")

# Load the BERT summarizer model
bert_model = Summarizer()

# Load the MBART model and tokenizer
translation_model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")
translation_tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")

LANGUAGE_CODES = {
    "Arabic": "ar_AR", "Czech": "cs_CZ", "German": "de_DE", "English": "en_XX",
    "Spanish": "es_XX", "Estonian": "et_EE", "Finnish": "fi_FI", "French": "fr_XX",
    "Gujarati": "gu_IN", "Hindi": "hi_IN", "Italian": "it_IT", "Japanese": "ja_XX",
    "Kazakh": "kk_KZ", "Korean": "ko_KR", "Lithuanian": "lt_LT", "Latvian": "lv_LV",
    "Burmese": "my_MM", "Nepali": "ne_NP", "Dutch": "nl_XX", "Romanian": "ro_RO",
    "Russian": "ru_RU", "Sinhala": "si_LK", "Turkish": "tr_TR", "Vietnamese": "vi_VN",
    "Chinese": "zh_CN", "Afrikaans": "af_ZA", "Azerbaijani": "az_AZ", "Bengali": "bn_IN",
    "Persian": "fa_IR", "Hebrew": "he_IL", "Croatian": "hr_HR", "Indonesian": "id_ID",
    "Georgian": "ka_GE", "Khmer": "km_KH", "Macedonian": "mk_MK", "Malayalam": "ml_IN",
    "Mongolian": "mn_MN", "Marathi": "mr_IN", "Polish": "pl_PL", "Pashto": "ps_AF",
    "Portuguese": "pt_XX", "Swedish": "sv_SE", "Swahili": "sw_KE", "Tamil": "ta_IN",
    "Telugu": "te_IN", "Thai": "th_TH", "Tagalog": "tl_XX", "Ukrainian": "uk_UA",
    "Urdu": "ur_PK", "Xhosa": "xh_ZA", "Galician": "gl_ES", "Slovene": "sl_SI"
}

def extract_audio_from_video(video_path, audio_path):
    # Load the video file
    video = mp.VideoFileClip(video_path)
    # Extract the audio
    video.audio.write_audiofile(audio_path)

def load_audio(file_path):
    return whisper.load_audio(file_path)

def split_audio(audio, chunk_length=30):
    sample_rate = whisper.audio.SAMPLE_RATE
    chunk_size = chunk_length * sample_rate
    chunks = [audio[i:i + chunk_size] for i in range(0, len(audio), chunk_size)]
    return chunks

def transcribe_chunk(model, audio_chunk):
    audio_chunk = whisper.pad_or_trim(audio_chunk)
    mel = whisper.log_mel_spectrogram(audio_chunk).to(model.device)
    _, probs = model.detect_language(mel)
    print(f"Detected language: {max(probs, key=probs.get)}")
    options = whisper.DecodingOptions()
    result = whisper.decode(model, mel, options)
    return result.text

def transcribe_audio(file_path):
    audio = load_audio(file_path)
    audio_chunks = split_audio(audio)
    transcription = ""

    for chunk in audio_chunks:
        transcription += transcribe_chunk(whisper_model, chunk) + " "

    return transcription.strip()

def transcribe_file(file_path):
    # Determine the file type based on extension
    _, file_extension = os.path.splitext(file_path)
    audio_path = "extracted_audio.wav"

    if file_extension.lower() in ['.mp4', '.avi', '.mov', '.mkv']:
        # It's a video file
        extract_audio_from_video(file_path, audio_path)
        transcription = transcribe_audio(audio_path)
    elif file_extension.lower() in ['.wav', '.mp3', '.flac', '.aac']:
        # It's an audio file
        transcription = transcribe_audio(file_path)
    else:
        raise ValueError("Unsupported file type. Please provide a video or audio file.")

    return transcription

def summarize_text(text):
    summary = bert_model(text, ratio=0.3)
    return summary

def translate_text(text, target_language):
    target_lang_code = LANGUAGE_CODES.get(target_language, "en_XX")
    model_inputs = translation_tokenizer(text, return_tensors="pt")
    generated_tokens = translation_model.generate(
        **model_inputs,
        forced_bos_token_id=translation_tokenizer.lang_code_to_id[target_lang_code]
    )
    translated_text = translation_tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    return translated_text[0]

def transcribe_summarize_translate(file, target_language):
    transcription = transcribe_file(file.name)
    summary = summarize_text(transcription)
    translation = translate_text(summary, target_language)
    return transcription, summary, translation

# Create the Gradio interface
interface = gr.Interface(
    fn=transcribe_summarize_translate,
    inputs=[gr.File(label="Upload an audio or video file"), gr.Dropdown(list(LANGUAGE_CODES.keys()), label="Select Target Language")],
    outputs=[gr.Textbox(label="Transcription"), gr.Textbox(label="Summary"), gr.Textbox(label="Translation")],
    title="Whisper Transcription, Summarization, and Translation",
    description="Upload an audio or video file to get the transcription, summary, and translation."
)

# Launch the Gradio interface
interface.launch()
```
